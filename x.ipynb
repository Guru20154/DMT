```python
nested_list = [[[3, 5, 7], [1, 2, 3]], [[3, 5, 7]]]
n = 2

result = []

# Iterate over the elements of the nested list
for sublist1 in nested_list:
    temp1 = []
    for sublist2 in sublist1:
        temp2 = [element / n for element in sublist2]
        print(temp2)
        temp1.append(temp2)
    result.append(temp1)

print(result)
```

```python
list1 = [[[0 for _ in range(3)]] for _ in range(10)]
list2 = [[2, 3, 4], [1, 2, 3]]
result = []

# Determine the maximum number of rows and columns between the two lists
max_rows = max(len(list1[0]), len(list2))
max_cols = 3#max(len(row) for row in list1[0] + list2)

# Iterate over the rows
for row in range(max_rows):
    temp = []

    # Iterate over the columns within each row
    for col in range(max_cols):
        # Add the corresponding elements if available, or use 0 otherwise
        element1 = list1[0][row][col] if row < len(list1[0]) and col < len(list1[0][row]) else 0
        element2 = list2[row][col] if row < len(list2) and col < len(list2[row]) else 0
        temp.append(element1 + element2)

    result.append(temp)

list1[0]=result
print(list1)
print(result)
# Output: [[3, 5, 7], [1, 2, 3]]
```

```python
import librosa
import numpy as np

audio_file = 'training-spectrograms/0_george_8.wav'
target_sr = 22050  # Set your desired target sample rate

audio, sr = librosa.load(audio_file)
audio_resampled = librosa.resample(y=audio,orig_sr= sr, target_sr=target_sr)

spectrogram = np.abs(librosa.stft(audio_resampled))
spectrogram.shape
```
